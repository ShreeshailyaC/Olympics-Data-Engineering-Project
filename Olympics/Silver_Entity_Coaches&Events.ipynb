## Dynamic Data Reading 

```python
dbutils.widgets.text("source_container","")
dbutils.widgets.text("sink_container","")
dbutils.widgets.text("folder","")
```

```python
#fetching parameters
source_container = dbutils.widgets.get("source_container")
sink_container = dbutils.widgets.get("sink_container")
folder = dbutils.widgets.get("folder")
```

```python
df = spark.read.format("parquet")\
    .load(f"abfss://{source_container}@olympicsprojectstorage.dfs.core.windows.net/{folder}")
```

```python
df.display()
```

```python
df.write.format("delta")\
    .mode("append")\
    .option("path",f"abfss://{sink_container}@olympicsprojectstorage.dfs.core.windows.net/{folder}")\
    .saveAsTable(f"olympicscatalog.{sink_container}.{folder}")
```
